```python
# Concatenate the three dataframes 
```

```python
df = pd.concat([file1, file2, file3], axis = 0)
df = df.reset_index(drop = True)
df.info() 
```

```python
#Delete the column education and the number of open complaints from the dataframe.
df = df.drop_duplicates()
df = df.drop(['education','number of open complaints'], axis=1) # axis = 1 -> columns
display(df.head())
```

```python
#Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
```

```python
def clean(x):
    if str(x)[-1]=='%':
        x = round(float(str(x).replace('%','')),4)/100
    else:
        x = round(float(x),2)
    return x
df['customer lifetime value']=df['customer lifetime value'].apply(clean)
display(df.head())
df.info()
```
